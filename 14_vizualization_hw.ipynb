```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
life = pd.read_csv('Life.csv', sep=';')
```

```python
life.info()
```

```python
life['PAYMENT MODE'].unique()
```

```python
life.Premium = life.Premium.str.replace(',', '').str.replace('-', '0')
```

```python
def issue_year(data):
    return int('20' + data['Issue Date'][4:6])
def premium_to_int(data):
    return int(data.Premium)
```

```python
life['Issue year'] = life.apply(issue_year, axis=1)
life['Premium'] = life.apply(premium_to_int, axis=1)
```

```python
life_years = life.groupby('Issue year').agg({'Premium': ['sum', 'count', 'mean']}) #.unstack()
life_years
```

```python
# выбран самый простой график, чтобы посмотреть, как ведет себя премия по годам, а также кол-во заключаемых договоров.
# График простой, неперегруженный. Видно, что премия росла до 2019 года, при этом кол-во падало => рост средней премии.
ax1 = life_years.Premium['sum'].plot()
ax2 = ax1.twinx()
ax2 = life_years.Premium['count'].plot(color='green')
```

```python
# посмотреть отдельно среднюю премию по годам. Хоть средняя премия и растет, такой рост не компенсирует падение в штуках =>
# падение выручки.
life_years.Premium['mean'].plot()
```

```python
life_pivot = (life.pivot_table(index = 'Issue year', columns = 'POLICY STATUS', values = 'Premium',
                               aggfunc = 'count', fill_value = 0, margins = True).reset_index().loc[0:8])
life_pivot['Share of lapsed'] = ((life_pivot['Lapse'] + life_pivot['Surrender'])/ life_pivot['All']) * 100
life_pivot
```

```python
# выбран этот график, чтобы посмотреть, как ведет себя процент расторгнутых договоров в зависимости от года заключения

plt.barh(life_pivot['Issue year'], life_pivot['Share of lapsed'])
plt.xlabel('% Lapsed')
plt.ylabel('Years')
plt.title('% of policies lapses by inception year')
```
